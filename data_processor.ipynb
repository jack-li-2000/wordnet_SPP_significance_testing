## Word Sense Disambiguation (WSD) Algorithm
By path distance. Pick the two synsets with the shortest path. 
Works well for same PoS pairs, but not very well for collocation pairs.

```python
def wsd(word1: str, word2: str) -> tuple[Synset, Synset]:
    synsets1: list[Synset] = wn.synsets(word1)
    synsets2: list[Synset] = wn.synsets(word2)

    max_sim_score: float = -1
    best_synsets: tuple[Synset, Synset] = (None, None)

    for s1 in synsets1:
        for s2 in synsets2:
            sim_score = s1.path_similarity(s2)
            if sim_score and sim_score > max_sim_score:
                max_sim_score = sim_score
                best_synsets = (s1, s2)

    return best_synsets
```

Apply and store WSD lookup table for unique pairs

```python
def count_unique_rows(df: DataFrame, *cols: str) -> DataFrame:
    return df[list(cols)].groupby(list(cols), as_index=False).value_counts()
```

```python
count_unique_rows(ntdf, "prime", "target")
```

```python
sense_pairs: dict[str, tuple[Synset, Synset]] = dict()


def build_lookup_table(row: Series) -> None:
    word_pair: tuple[str, str] = (row["prime"],  row["target"])
    if (sense_pair := wsd(*word_pair))[0] is not None:
        sense_pairs[",".join(word_pair)] = sense_pair


_ = count_unique_rows(ntdf, "prime", "target").apply(
    build_lookup_table, axis=1)

sense_pairs
```

```python
def compute_similarity_from_wordnet(row: Series) -> Series:
    word_pair: str = row["prime"] + "," + row["target"]
    sims = [np.nan] * 6
    if word_pair in sense_pairs:
        prime_synset, target_synset = sense_pairs[word_pair]
        for i in range(6):
            try:
                if i < 3:
                    sim = SIM_FUNCTIONS[i](prime_synset, target_synset)
                else:
                    sim = getattr(target_synset, SIM_FUNCTIONS[i])(
                        prime_synset, IC)
            except:
                pass
                # sims[i] = np.nan
            else:
                sims[i] = sim

    return pd.Series(sims)


test = ntdf[:100].copy()
test[SIM_NAMES] = test.apply(compute_similarity_from_wordnet, axis=1)

test
```

# Collocation Ratings

```python
# from nltk.probability import ConditionalFreqDist, ConditionalProbDist, MLEProbDist
from nltk.probability import FreqDist
from nltk.corpus import reuters, brown, gutenberg, webtext

nltk.download('reuters', download_dir="./nltk_data")
nltk.download('brown', download_dir="./nltk_data")
nltk.download('gutenberg', download_dir="./nltk_data")
nltk.download('webtext', download_dir="./nltk_data")
```

## Corpora and bigram model

```python
class BigramModel:
    def __init__(self, corpus_text: nltk.Text):
        bigrams = list(nltk.bigrams(corpus_text))
        self.bigram_counts = FreqDist(bigrams)

    def next_word_freqs(self, prime: str, target: str) -> tuple[int, int]:

        bigrams_starting_with_prime_count: int = 0
        target_count: int = 0

        for bigram, count in self.bigram_counts.items():
            if bigram[0] == prime:
                bigrams_starting_with_prime_count += count
                if bigram[1] == target:
                    target_count = count

        return target_count, bigrams_starting_with_prime_count
```

```python
corpora = (reuters, brown, gutenberg, webtext)
CORPORA_NAMES = ["reuters", "brown", "gutenberg", "webtext", "weighted_avg"]
BIGRAM_MODELS: list[BigramModel] = list(BigramModel(nltk.Text(corpus.words())) for corpus in corpora)
```

## Compute collocation ratings
run computation and save it to `csv`, 
or load from saved `csv` from last running, since last time it takes 550 minutes to run on the whole 0.4 million datapoints.

```python
def compute_collocation_ratings_from_corpora(row: Series) -> Series:
    prime: str = row["prime"]
    target: str = row["target"]

    target_counts: int = 0
    prime_total_counts: int = 0

    ratings: list[float] = []

    for model in BIGRAM_MODELS:
        model: BigramModel
        count, total = model.next_word_freqs(prime, target)
        if total == 0:
            ratings.append(np.nan)
            continue
        ratings.append(count / total)
        target_counts += count
        prime_total_counts += total

    ratings.append(np.nan if prime_total_counts == 0
                   else target_counts / prime_total_counts)

    return pd.Series(ratings)


test = ntdf[:10].copy()
test[CORPORA_NAMES] = test.apply(
    compute_collocation_ratings_from_corpora, axis=1)

test
```

```python
# load if file saved instead of recomputing
if os.path.exists(csv_fn := "./data/NT_trials_full_table.csv"):
    ntdf = pd.read_csv(csv_fn)
else:
    ntdf[CORPORA_NAMES] = ntdf.apply(compute_collocation_ratings_from_corpora, axis=1)
    ntdf.to_csv(csv_fn, index=False)
```
